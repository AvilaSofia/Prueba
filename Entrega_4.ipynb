```python
import numpy as np
import pandas as pd
import yfinance as yf
from arch import arch_model
from scipy.stats import norm
import matplotlib.pyplot as plt

# Datos del portafolio
tickers = ['CRM','UNH','CVX','MSFT']
weights = np.array([0.000621, 0.418453, 0.479852, 0.153705])
start = '2025-09-11'

# Descargar precios
prices = yf.download(tickers, start=start, progress=False)['Close']
rets = prices.pct_change().dropna()

# Rendimientos del portafolio
port_daily = rets.dot(weights)
port_daily.name = 'Rendimientos del Portafolio'
```

```python
plt.figure(figsize=(10,5))
port_daily.plot(title='Rendimientos diarios del portafolio')
plt.show()
```

```python
# ARCH(1)
arch_1 = arch_model(port_daily*100, mean='constant', vol='ARCH', p=1)
res_arch_1 = arch_1.fit(disp='off')
print(res_arch_1.summary())

# Varianza condicional
res_arch_1.plot(annualize='D')
plt.suptitle('Varianza condicional  Modelo ARCH(1)', fontsize=12)
plt.show()
```

```python
# ARCH(2)
arch_2 = arch_model(port_daily*100, mean='constant', vol='ARCH', p=2)
res_arch_2 = arch_2.fit(disp='off')
print(res_arch_2.summary())

# Varianza condicional
res_arch_2.plot(annualize='D')
plt.suptitle('Varianza condicional  Modelo ARCH(2)', fontsize=12)
plt.show()
```

```python
# GARCH(1,1)
garch_11 = arch_model(port_daily*100, mean='constant', vol='GARCH', p=1, q=1)
res_garch_11 = garch_11.fit(disp='off')
print(res_garch_11.summary())

# Varianza condicional
res_garch_11.plot(annualize='D')
plt.suptitle('Varianza condicional  Modelo GARCH(1,1)', fontsize=12)
plt.show()
```

```python
# GARCH(1,2)
garch_12 = arch_model(port_daily*100, mean='constant', vol='GARCH', p=1, q=2)
res_garch_12 = garch_12.fit(disp='off')
print(res_garch_12.summary())

# Varianza condicional
res_garch_12.plot(annualize='D')
plt.suptitle('Varianza condicional  Modelo GARCH(1,2)', fontsize=12)
plt.show()
```

```python
# GARCH(2,1)
garch_21 = arch_model(port_daily*100, mean='constant', vol='GARCH', p=2, q=1)
res_garch_21 = garch_21.fit(disp='off')
print(res_garch_21.summary())

# Varianza condicional
res_garch_21.plot(annualize='D')
plt.suptitle('Varianza condicional  Modelo GARCH(2,1)', fontsize=12)
plt.show()
```

# VaR GARCH (1,1)

```python
def calcular_var_multi(res, horizontes=(1, 21, 252), niveles=(0.95, 0.99)):
    resultados = {}
    for H in horizontes:
        forecast = res.forecast(horizon=H, reindex=False)
        mu = forecast.mean.values[-1, :H] / 100
        var = forecast.variance.values[-1, :H] / (100**2)
        mu_H = mu.sum()
        sigma_H = np.sqrt(var.sum())
        for cl in niveles:
            key = f'VaR_{int(cl*100)}_{H}d'
            resultados[key] = -(mu_H + sigma_H * norm.ppf(1 - cl))
    return pd.DataFrame(resultados, index=['Valor'])

VaR_multi = calcular_var_multi(res_garch_11, horizontes=(1, 21, 252), niveles=(0.95, 0.99))
print(VaR_multi)

sigma_t = res_garch_11.conditional_volatility / 100
mu_t = pd.Series(res_garch_11.params.get('mu', 0)/100, index=sigma_t.index)

VaR95_d = -(mu_t + sigma_t * norm.ppf(0.05))
VaR99_d = -(mu_t + sigma_t * norm.ppf(0.01))

vaR_insample = pd.DataFrame({
    'Ret_Port': port_daily,
    'VaR_95_1d': VaR95_d,
    'VaR_99_1d': VaR99_d
}).dropna()

plt.figure(figsize=(12, 5))
plt.plot(vaR_insample.index, vaR_insample['Ret_Port'], label='Rendimientos diarios')
plt.plot(vaR_insample.index, -vaR_insample['VaR_95_1d'], label='-VaR 95% (1D)')
plt.plot(vaR_insample.index, -vaR_insample['VaR_99_1d'], label='-VaR 99% (1D)')

viol_95 = vaR_insample['Ret_Port'] < -vaR_insample['VaR_95_1d']
viol_99 = vaR_insample['Ret_Port'] < -vaR_insample['VaR_99_1d']

plt.scatter(vaR_insample.index[viol_95],
            vaR_insample.loc[viol_95, 'Ret_Port'],
            s=18, color='red', label='Violaciones 95%')
plt.scatter(vaR_insample.index[viol_99],
            vaR_insample.loc[viol_99, 'Ret_Port'],
            s=28, color='orange', label='Violaciones 99%')

plt.title('VaR Diario (GARCH(1,1)) con Violaciones')
plt.xlabel('Fecha')
plt.ylabel('Retorno')
plt.legend()
plt.tight_layout()
plt.show()
```
